## 预测模型之——线性回归(Linear Regression)

线性回归：

- 基本的预测模型之一；
- 估计了变量之间的关系；
- 假定被预测变量属于正态分布
- 假定残差属于正态分布
- 假定预测变量之间互相独立


优点：

- 简单、易于理解
- 常用于变量之间的线性关系

缺点：

- 很难满足各种假定条件
- 易受异常值的影响

为什么叫“回归”呢？

这个词来源于19世纪统计学家Galton所描述的一种生物现象，他收集了父子两代人的身高，观察到：

- 高个子的后代往往身高会低点，矮个子的后代往往身高会高点，都趋向于平均身高，即**回归于平均身高**。
- 否则，如果高个子的后代身高更高，矮个子的后代身高更低，一代代这样发展下来，人类的身高会趋向于两个极端，而不是现在这样，身高差不多的现象。

我们先看下简单的一元线性回归，直观的感受下：

假定我们有10个婴儿的出生月份和体重的数据，我们想找出婴儿出生月份和体重之间的关系，然后根据婴儿的出生月份预测体重。

```python
infant = pd.DataFrame({'months':[1, 3, 5, 2, 11, 9, 3, 9, 12, 3], 'weight':[4.4, 5.3, 7.2, 5.2, 8.5, 7.3, 6.0, 10.4, 10.2, 6.1]})
```

```python
infant.head()
```

```python
plt.scatter(infant.months, infant.weight)
plt.xlabel('months')
plt.ylabel('weight')
plt.title(u'婴儿体重与出生月份的散点图');
```

我们构造如下关系，x代表出生月份，y代表婴儿体重，随着出生月份的增长，婴儿体重也随之增加。

![text](img/regression_function.PNG)

上面数学公式描述的，就是所谓的**回归线**，满足条件的直线应该有很多条，哪条是最好的呢？

我们根据**普通最小二乘法(ordinary least square)**来选择，即预测值与真实值之间的距离平方和最小的直线。

```python
plt.scatter(infant.months, infant.weight)
plt.plot(np.linspace(0, 13, 10), 4.0 + 0.5* np.linspace(0, 13, 10), 'r--')
plt.plot(np.linspace(0, 13, 10), 3.0 + 0.7* np.linspace(0, 13, 10), 'g--')
plt.xlabel('months')
plt.ylabel('weight')
plt.title(u'婴儿体重与出生月份的关系')
plt.xlim(0, 14)
plt.ylim(4, 11);
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
# split the data 
x_train, x_test, y_train, y_test = train_test_split(infant.months, infant.weight, test_size = 0.2, random_state = 0)
```

```python
reg = LinearRegression()
```

```python
# train the model 
reg.fit(x_train[:, np.newaxis], y_train)
```

模型拟合的回归线的截距和斜率。

```python
reg.intercept_
```

```python
reg.coef_
```

拟合优度：回归直线对观测值的拟合程度。

度量拟合优度的统计量时R平方，范围在0和1之间，1代表拟合优度高，0代表拟合的差。

```python
r2_score(y_test, reg.predict(x_test[:, np.newaxis]))
```

另外一个反映模型对观测值的拟合好坏的指标——均方误差（Mean Square Error）

![mse](img/regression_mse.PNG)

即真实值与预测值的差的平方，再计算均值。

**均方误差越小，模型的效果越好。**

```python
mean_squared_error(y_test, reg.predict(x_test[:, np.newaxis]))
```

```python
plt.scatter(infant.months, infant.weight)
plt.plot(infant.months, reg.predict(infant.months[:, np.newaxis]), 'r--')
plt.xlabel('months')
plt.ylabel('weight')
plt.title(u'婴儿体重与出生月份的散点图与回归线');
```

下图可以看到，图中以高中生的体重和身高为例，用直线近似两者之间的关系，但当数据包括了婴儿及成年人的数据后，整体的关系，需要用曲线来拟合，而不是直线。

## 预测模型之——回归模型在高维的衍生模型

前面的婴儿的示例数据，有10个观测值，有一个解释变量，一个被预测变量，这种观测值个数>解释变量个数的，通常称为低维数据。

观测值个数<解释变量个数的，称为高维数据，以基因数据最典型，人类大约有2-3万个基因，如果一个基因表示一个维度，那就是2万到3万个维度。

对于这样的数据，线性回归方程是无解的，但是在线性回归方程的基础上，衍生出了针对高维数据的回归模型。

由于解释变量个数太多，才导致线性回归方程无解，因此，对变量的系数进行限制（对系数的正则化regularization），引出了新的模型。

在计算一元线性回归线时，我们提到了普通最小二乘法，其数学公式如下所示：

![ols](img/ols.PNG)

**岭回归(ridge regression)**

- 添加了对变量系数的模的平方的限制，这个限制有助于消除解释变量之间的共线性，其数学公式如下所示：
![ridge](img/ridge.PNG)
- 伴随调节参数的变化，变量系数也随之变化，当调节参数为0时，等同于一般的线性回归，见下图：
![ridge_p](img/ridge_plot.PNG)

**Lasso(least absolute shrinkage and selection operator)**

- 添加了对变量系数的模的**绝对值**的限制，可以进行**变量选择**，其数学公式如下所示：
![lasso](img/lasso.PNG)

**弹性网(Elastic Net)**

- 上面两个模型的混合体，其数学公式如下所示：
![net](img/net.PNG)
- 伴随调节参数的变化，变量系数也随之变化，lasso和elastic net的系数变化，见下图：
![net](img/net_path.PNG)

类似的算法还有：贝叶斯岭回归、逻辑回归的penalty参数、Least-angle regression (LARS)。
针对线性回归对异常值敏感的缺点，有稳健回归(Robustness regression)；
针对线性回归的线性关系，有多项式回归构建非线性关系。

**下面以糖尿病数据为例，来看下回归模型的使用：**

```python
from sklearn.datasets import load_diabetes
from sklearn.datasets import load_boston

from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV, LarsCV, LassoLarsCV

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
diabetes = load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.data.shape
```

```python
pd.DataFrame(diabetes.data).plot(subplots=True);
```

```python
pd.Series(diabetes.target).plot();
```

```python
diabetes.data[:5, :]
```

```python
diabetes.target[:5]
```

**该数据集包含442个病人的测量而得的10项生理指标(年龄，性别，体重，血压)，和一年后疾病进展的数据。**

**这个任务是用来从生理指标预测疾病。**

**这里不仅使用回归模型进行预测，部分前面讲过的分类模型，也可以用做回归预测。**

**这里使用三个指标——R方、均方误差、均绝对值误差，来评价各预测模型的效果。**

```python
# model
lr = LinearRegression()
ridge = RidgeCV()
lasso = LassoCV()
net = ElasticNetCV()
lar = LarsCV()
lassolars = LassoLarsCV()

tree = DecisionTreeRegressor(random_state = 0)
knn = KNeighborsRegressor()
svr_l = SVR(kernel = 'linear')
svr_r = SVR()
svr_p = SVR(kernel = 'poly')

n_estimators = 100
bag = BaggingRegressor(DecisionTreeRegressor(), n_estimators = n_estimators, random_state = 0)
rf = RandomForestRegressor(n_estimators = n_estimators, random_state = 0)
etree = ExtraTreesRegressor(n_estimators = n_estimators, random_state = 0)
ada = AdaBoostRegressor(DecisionTreeRegressor(), n_estimators = n_estimators, random_state = 0)
gbr = GradientBoostingRegressor(n_estimators = n_estimators, random_state = 0)

model_list = [lr, ridge, lasso, net, lar, lassolars, tree, knn, svr_l, svr_r, svr_p, bag, rf, etree, ada, gbr]
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_squared_error')
    score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'r2')
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_absolute_error')
    score_mean.append(score.mean())
    score_std.append(score.std())
```

```python
plt.errorbar(range(16), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 16)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'svr_p', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('R2')
plt.title(u'各模型预测糖尿病的性能');
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_squared_error')
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'r2')
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_absolute_error')
    score_mean.append(score.mean())
    score_std.append(score.std())
```

```python
plt.errorbar(range(16), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 16)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'svr_p', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('mean_square_error')
plt.title(u'各模型预测糖尿病的性能');
```

```python
score_mean = list()
score_std = list()
for model in model_list:
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_squared_error')
    #score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'r2')
    score = cross_val_score(model, diabetes.data, diabetes.target, scoring = 'mean_absolute_error')
    score_mean.append(score.mean())
    score_std.append(score.std())
```

```python
plt.errorbar(range(16), score_mean, yerr=score_std, fmt = 'ob')
plt.xlim(-1, 16)
plt.xticks(range(16), ('lr', 'ridge', 'lasso', 'net', 'lar', 'lalar', 'tree', 'knn', 'svr_l', 'svr_r', 'svr_p', 'bag', 'rf', 'ETree', 'ada', 'gb'))
plt.xlabel('Models')
plt.ylabel('mean_absolute_error')
plt.title(u'各模型预测糖尿病的性能');
```

从上图可以看到，数据的前后结构差异较大。

- 如果抽取训练集和测试集时，刚好把前面有规律的数据划分为训练集，后面结构突变的部分划分为测试集，则不管什么模型的效果都会突然变差；
- 把数据打乱，**随机抽取数据很重要，保证了训练集和测试集中数据结构一致性**；
- 这也是为什么要数据越多越好，包括的特殊的结构就多，训练模型时，就可以更好的应对。
