 **预测全部正确！**
 
 真的吗？
 
 有可能是假的。
 
 这貌似就是传说中的——**过拟合（outfit）**。
 
 顾名思义，拟合模型过度了。
 
 - 因为我们用全部的数据去训练模型，又直接用老数据去预测；
 - 即该模型训练时知道正确的类别，然后千方百计构造规则，得到正确类别。
 
 那怎么办呢？
 
 把数据拆分为两份，一份专门训练模型（训练集），另外一份用来预测，测试模型（验证集）。
 
 - 这个方法有个缺点：由于是随机的把数据拆分为两份，所以最后验证集分类准确率的高低与原始数据的分组有很大的关系。
 - 比如：setosa类别的数据恰好被选为训练集，训练出的模型规则会比较简单；
 - 非setosa类别的数据恰好被选为训练集，训练出的模型规则会较复杂；
 
 较流行的是**k重交叉验证（k-fold cross validation）**;
 - 将数据拆分为k组，使用每一个子集数据分别做一次验证集，其余的k-1组子数据做训练集，这样得到k个模型，用这k个模型的最终的验证集的分类准确率的平均值作为此k重交叉验证下分类器的性能指标。

### 决策树——k重交叉验证

````python
from sklearn.cross_validation import KFold
````

````python
kf = KFold(n = 150, n_folds = 3, shuffle = False)
````

````python
index = list()
for train_index, test_index in kf:
    index.append('test: %s train: %s' % (test_index, train_index))    
````

````python
index
````

上面的拆分有点小问题：

- 第一个模型的验证集数据都是种类为0的，第二个模型的验证数据都是种类为1的；
- 即第一个模型训练时没见过种类0的数据，第二个模型没见过种类为1的数据。

因此，在做分类模型时，最好使用k重交叉验证的衍生版本**StratifiedKFold**
- 拆分数据时考虑了y变量的分类信息，避免上面的不平衡的类别分布。

````python
from sklearn.cross_validation import StratifiedKFold
````

````python
skf = StratifiedKFold(y = iris.target, n_folds = 3, shuffle = False)
````

````python
index_s = list()
for train_index, test_index in skf:
    index_s.append('test: %s train: %s' % (test_index, train_index))    
````

````python
index_s
````

````python
skf = StratifiedKFold(y = iris.target, n_folds = 3, shuffle = False)
scores = []
for train_index, test_index in skf:
    # split the data 
    x_train_s, x_test_s = iris.data[train_index], iris.data[test_index]
    y_train_s, y_test_s = iris.target[train_index], iris.target[test_index]
    # train and test the model
    scores.append(accuracy_score(y_test_s, tree.DecisionTreeClassifier().fit(x_train_s, y_train_s).predict(x_test_s)))
````

````python
scores_1 = pd.Series(scores)
scores_1
````

### 决策树——分类模型的各种评价指标

针对分类模型的评价指标，只有上面计算的准确度（Accuracy）是远远不够的，比如：

- 你有一个银行的信用数据，假设有100个客户申请了贷款，其中有5个客户欠款不还，其他95个客户都按期还贷。
- 要求你根据现有的这100个客户的资料和是否按期还贷，训练一个分类模型，根据新的客户资料，提前预测客户是否会按期还贷。
- 现在你训练了一个模型，准确度0.9（90%），正确识别出了3个欠款不还的客户，87个按期还贷的客户，其余的10个客户中，2个欠款不还的客户被误分为好客户，8个按期还贷的客户被误分为坏客户。
- 有人说了，我不用模型，直接把100个客户都分类为好客户，准确度95%，比你的模型效果还好！

真的是这样吗？
- 一个坏客户给银行带来的损失**远超过**一个好客户给银行带来的收益；
- 在保证一定的正确率的前提下，关键是识别出坏客户。
- 我们就需要其他的指标来告诉我们，坏客户有多少被识别出来？好客户有多少被误判为坏客户？等等

这些指标的获得都要从**分类矩阵（混淆矩阵 confusion matrix）**开始。

````python
from sklearn.metrics import confusion_matrix
````

````python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 123)
# train and predict model
clf_tree = tree.DecisionTreeClassifier(random_state = 123)
y_pred = clf_tree.fit(x_train, y_train).predict(x_test)
````

````python
clf_tree
````

````python
def plot_confusion_matrix(cm, title = 'Confusion matrix', cmap = plt.cm.Blues):
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(iris.target_names))
    plt.xticks(tick_marks, iris.target_names, rotation = 45)
    plt.yticks(tick_marks, iris.target_names)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
````

````python
cm = confusion_matrix(y_test, y_pred)
cm
````

````python
plt.figure()
plot_confusion_matrix(cm)
````

上面的混淆矩阵：

- 对角线上的元素都是分类正确的；
- 非对角线上的元素都是分类错误的；
- 颜色越深，代表数值越大，如对角线上第一个元素，setosa对应的方块数值为13，颜色最深。

看上面的图有可能会误导：

- setosa类别分类正确的数值多于versicolor，确实13>6；
- 不过，我们更想看到的是**分类正确的比率**，实际上，setosa类别和versicolor类别都被正确分类。
- 不同在于，验证集中setosa类别的数据有13个，versicolor类别的数据有6个。
- 我们可以对混淆矩阵标准化，每个数值都除以验证集中所属真实类别的总数量。
- 比如13/13，6/6，1/11，10/11。


````python
# Normalize the confusion matrix by row (i.e by the number of samples in each class)
cm_normalized = cm.astype('float') / cm.sum(axis = 1)[:, np.newaxis]
cm_normalized
````

````python
plt.figure()
plot_confusion_matrix(cm_normalized, title = 'Normalized confusion matrix')
````

其中，较常用的是precision（正确率）和recall（召回率）。

- False Positive rate：真正的坏客户的人群中，被误判为好客户的占比。 
- Specificity：      真正的坏客户的人群中，被预测为坏客户的占比。
- precision：       预测为好客户的人群中，真正好客户的占比。
- recall（True Positive rate/ sensitivity）  ： 真正的好客户的人群中，被预测为好客户的占比。

这几个指标的含义，真的很容易混淆，怪不得这矩阵叫**混淆矩阵**，名副其实啊。

我们先不管这几个指标的具体含义，先看下他们的来源：

- 一个混淆矩阵，能计算一个上面的precision值，一个recall值等。
- 混淆矩阵是怎么来的？
- 是预测值和真实值比较计算来的
- 真实值是固定不变的，不用考虑
- 预测值怎么来的？
- 决策树直接算出来的
- 决策树怎么算的？通过概率计算的。
- 不过简化起见，我们直接用了预测值，没用概率值；
- 其实就是预设了阈值，假设为0.5，则概率值大于等于0.5的均分类为1，概率值小于0.5的均分类为0。
- 试想，如果我们设置阈值为0.4，不是又得到一个新的混淆矩阵，又得到一个新的precision值，一个新的recall值等。
- 我们的阈值不断的改变，就能得到一系列的precision和一系列的recall值等。
- 随着阈值的变化，precision和recall等值就会随之变化。
- 那么，我们要动态的评价，一个合理的办法就是把基于不同的阈值而产生的一系列指标描绘到直角坐标系上，就能清楚的看到他们的对应关系。
- 我们用FP rate做横轴，recall做纵轴，就产生了著名的**ROC曲线**（receiver operating characteristic curve，又称为接受者操作特性曲线）

````python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp
````

````python
# Binarize the output
y_roc = label_binarize(iris.target, classes = [0, 1, 2])
n_classes = y_roc.shape[1]
````

````python
# add noisy features to make the problem harder
random_state = np.random.RandomState(0)
n_samples, n_features = iris.data.shape
x_roc = np.c_[iris.data, random_state.randn(n_samples, 200*n_features)]
````

````python
# shuffle and split traning and test sets
x_train_roc, x_test_roc, y_train_roc, y_test_roc = train_test_split(x_roc, y_roc, 
                                                                    test_size = .5, 
                                                                    random_state = 0)
````

````python
# learn to predict each class against the other
classifier = OneVsRestClassifier(tree.DecisionTreeClassifier(random_state = random_state))
y_score_roc = classifier.fit(x_train_roc, y_train_roc).predict_proba(x_test_roc)
````

````python
y_score_roc[:5, :]
````

````python
# compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_roc[:, i], y_score_roc[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
````

````python
# compute micro-average ROC curve and ROC area
fpr['micro'], tpr['micro'], _ = roc_curve(y_test_roc.ravel(), y_score_roc.ravel())
roc_auc['micro'] = auc(fpr['micro'], tpr['micro'])
````

````python
# plot of a ROC curve for a specific class
plt.figure()
plt.plot(fpr[2], tpr[2], label = 'ROC curve (area = %0.2f)' % roc_auc[2])
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc = 'lower right');
````

````python
# plot ROC curves for the multiclass problem

# compute macro-average ROC curve and ROC area

# first aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

# then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += interp(all_fpr, fpr[i], tpr[i])
    
# finally average it and compute AUC
mean_tpr /= n_classes

fpr['macro'] = all_fpr
tpr['macro'] = mean_tpr
roc_auc['macro'] = auc(fpr['macro'], tpr['macro'])

# plot all ROC curves
plt.figure()
plt.plot(fpr['micro'], tpr['micro'], 
         label = 'micro-average ROC curve (area = %0.2f)' % roc_auc['micro'], 
         linewidth = 2)
plt.plot(fpr['macro'], tpr['macro'], 
         label = 'macro-average ROC curve (area = %0.2f)' % roc_auc['macro'], 
         linewidth = 2)

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label = 'ROC curve of class %d (area = %0.2f)' % (i, roc_auc[i]))


plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Some extension of Receiver operating characteristic to multi-classes')
plt.legend(loc = 'lower right');
````

**ROC曲线**：
```
- 以坏客户蒙混过关的概率为横轴，
- 好客户被准确识别的概率为纵轴，
- 在同一个模型中，由于采用不同的概率阈值得出的不同结果画出的曲线。

- 图中的实线，代表阈值变化时分类模型给出的TPR与FPR的变化，
- 图中的虚线，代表随机猜测的TPR与FPR的变化。
- 如果实线在虚线之上，代表你的模型还时比胡猜优秀，
- 如果实线在虚线之下，你可以放弃你的模型了。

- 一个理想的模型会尽可能的处于左上角，这意味着TPR要尽可能大，FPR尽可能小。
- ROC有一个更大的优点，可以基于成本效益来判决分类器好坏。
- 仍然以银行的信用数据为例，我们可以判断下图的A与B谁更优秀。
- 在贷款金额小，真客户带来的潜在收益大，扩张阶段，选B点。
- 在贷款金额大，坏客户的违约破坏大于真客户带来的实际收益，选A点。

- ROC曲线是根据与45度线的偏离来判断模型好坏。
- 图形的好处是直观，缺点是不够精确。
- 一个自然的想法是对曲线下的面积进行比较，
- 这就得到了AUC（area under the ROC curve，ROC曲线下的面积）。
```

- ROC曲线多用于二类别分类模型效果的评价；
- 我们这里鸢尾花的数据，需要判别三个类别。
- 第一种方法：二值化处理。
- 每次选一个类别，比如类别0，其他两个类别均为非0,画出单类别的ROC曲线；
- 第二种方法：把通过第一种方法计算的三个类别的FPR和TPR数值平均，画出汇总的ROC曲线。
- 平均又分两种方法，一个是把每一行数据作为一个单位，是否分类正确，求平均，得到micro平均曲线。
- 另一个是把一个类别（其中有50行数据）作为一个单位，是否分类正确，求平均，得到macro平均曲线。

不只是FPR和TPR两个指标可以构成一个曲线，**在推荐或信息获取领域，还会组合使用precision和recall(TPR)构造PR曲线**。

- PR曲线应该是来源于信息检索中对相关性的评价。
- 以上面的银行信用数据为例，precision就是决定发放贷款的人群中有多少会按时还款；
- recall，所有值得发放贷款的人，有多少被银行挑选出来，发放贷款了；
- 放到检索中来说，precision就是检索出来的条目中（比如网页）有多少是准确的的；
- recall，所有准确的条目有多少被检索出来了。

````python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
````

````python
# compute precision-recall and plot curve
precision = dict()
recall = dict()
average_precision = dict()
for i in range(n_classes):
    precision[i], recall[i], _ = precision_recall_curve(y_test_roc[:, i], 
                                                        y_score_roc[:, i])
    average_precision[i] = average_precision_score(y_test_roc[:, i], 
                                                   y_score_roc[:, i])
    
# compute micro-average ROC curve and ROC area
precision['micro'], recall['micro'], _ = precision_recall_curve(y_test_roc.ravel(), 
                                                                y_score_roc.ravel())
average_precision['micro'] = average_precision_score(y_test_roc, y_score_roc, 
                                                     average = 'micro')

# plot precision-recall curve
plt.figure()
plt.plot(recall[2], precision[2], label = 'Precision-Recall curve')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall example: AUC=%0.2f' % average_precision[2])
plt.legend(loc = 'lower right');
````

````python
# plot precision-recall curve for each class
plt.figure()
plt.plot(recall['micro'], precision['micro'], 
         label = 'micro-average Precision-recall curve (area = %0.2f)' % average_precision['micro'])

for i in range(n_classes):
    plt.plot(recall[i], precision[i], 
             label = 'Precision-recall curve of class %d (area = %0.2f)' % (i, average_precision[i]))

    
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Extension of Precision-Recall to multi-classes')
plt.legend(loc = 'lower right');
````
