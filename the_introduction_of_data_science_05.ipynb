## 聚类模型之——k-means

不管是分类模型，还是回归模型，都有Y变量的数据，这称为**有监督学习(supervised learning)**。

那如果没有Y变量呢？

这就是**无监督学习(unsupervised learning)**，包括聚类模型，降维模型如主成分分析、因子分析等。

聚类clustering，简单地说就是把相似的东西分到一组。

我们先看下聚类模型中的K-means聚类：

如下图所示，数据样本用圆点表示，每个簇的中心点用叉叉表示。

```python
Image('img/k-means.PNG')
```

- (a)刚开始时是原始数据，杂乱无章，没有label，看起来都一样，都是绿色的。
- (b)假设数据集可以分为两类，令K=2，随机在坐标上选两个点，作为两个类的中心点。
- (c-f)演示了聚类的两种迭代。
- 先划分，把每个数据样本划分到最近的中心点那一簇；
- 划分完后，更新每个簇的中心，即把该簇的所有数据点的坐标加起来求平均值。
- 这样不断进行”划分—更新—划分—更新”，直到每个簇的中心不在移动为止。

**优点**

- 简单、非参数

**缺点**

- 计算量大；维度的诅咒（提前用PCA降维，效果较好）；
- 受初始中心点选择的影响，有时找到的类别中心点是局部最优，不是全局最优；
- 易被噪音或不相关变量影响，降低准确性，特征选择和转换影响效能；
- 对数据的局部结构敏感，即类别的分布有偏，影响效果；
- 如银行的信用数据中，优质客户居多，来一个新记录，被分类为优质的概率就大。

k-means的衍生模型：

**k-medoids**，区别在于中心点的选取。

- k-means中选当前类别中所有数据点的平均值;
- 在 k-medoids 算法中，我们将从当前 cluster 中选取这样一个点——它到其他所有（当前 cluster 中的）点的距离之和最小——作为中心点。
- k-means 和 k-medoids 之间的差异就类似于一个数据样本的均值 (mean) 和中位数 (median) 之间的差异：前者的取值范围可以是连续空间中的任意值，而后者只能在给样本给定的那些点里面选。
- 优势：不易受异常值影响；
- 缺点：计算量大，仅适用于小数据集。

**GMM(Gaussian Mixture Model)**

- k-means的结果是每个数据点被分配到某一个类别。
- GMM则给出这些数据点被分配到每个类别的概率。
- 概率的信息量比简单的一个结果要多。比如，可以把概率转换为一个得分，表示算法对自己得出的这个结果的把握。
- 另一个很常见的方法是在诸如疾病诊断之类的场所，机器对于那些很容易分辨的情况（患病或者不患病的概率很高）可以自动区分，而对于那种很难分辨的情况，比如，49% 的概率患病，51% 的概率正常，如果仅仅简单地使用 50% 的阈值将患者诊断为“正常”的话，风险是非常大的，因此，在机器对自己的结果把握很小的情况下，会“拒绝发表评论”，而把这个任务留给有经验的医生去解决。

**层次聚类**
把数据表示成一棵树的形式，通过拆分树，来得到类别。

**降维的必要性**

1. 多重共线性--预测变量之间相互关联。多重共线性会导致解空间的不稳定，从而可能导致结果的不连贯。

2. 高维空间本身具有稀疏性。一维正态分布有68%的值落于正负标准差之间，而在十维空间上只有0.02%。 

3. 过多的变量会妨碍查找规律的建立。

4. 仅在变量层面上分析可能会忽略变量之间的潜在联系。例如几个预测变量可能落入仅反映数据某一方面特征的一个组内。

**降维的目的**

1. 减少预测变量的个数

2. 确保这些变量是相互独立的

3. 提供一个框架来解释结果

## 降维模型之——主成分分析PCA(Principal Component Analysis)

PCA原理：

- PCA把原先的n个特征用数目更少的m个特征取代，新特征是旧特征的线性组合，这些线性组合最大化样本方差，尽量使新的m个特征互不相关。
- 不仅仅是对高维数据进行降维，更重要的是经过降维去除了噪声，发现了数据中的模式。

**为什么要最大化样本方差呢？**

- 对于一系列特征组成的多维数据，多维数据里的某些变量本身没有区分性。
- 比如某个变量所有的数据都为1，或者与1差距不大，那么这个变量本身就没有区分性，用它做特征来区分，贡献会非常小。
- 所以我们的目的是找那些变化大的变量，即方差大的那些维，而去除掉那些变化不大的维，从而使特征留下的都是精品，而且计算量也变小了。

PCA过程：

1. 矩阵A特征中心化后得到矩阵B。即每一维的数据都减去该维的均值。这里的“维”指的就是一个特征（或属性），变换之后每一维的均值都变成了0。
2. 计算B的协方差矩阵C。
3. 计算协方差矩阵C的特征值和特征向量。
4. 选取大的特征值对应的特征向量，得到新的数据集。

PCA优点：

- 无参数。即不需要设定参数或根据任何经验模型对计算进行干预，最后的结果只与数据相关。

PCA缺点：

- 如果用户对观测对象有一定的先验知识，掌握了数据的一些特征，却无法通过参数化等方法对处理过程进行干预，可能不能达到预期的效果。

**PCA应用之一：以鸢尾花数据为例，计算PCA**

- 第一主成分解释的方差变动最多，使用第一主成分就可以对鸢尾花的类别做大致的分类；
- 第一主成分和第二主成分组合，方便查可视化数据的内部结构。

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# load the dataset
iris = load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

# PCA 
pca = PCA(n_components=2)
pca_x = pca.fit(iris.data).transform(iris.data)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

**PCA的应用之三：PCA + LogisticRegression**

以手写体数字识别为例：

- 数字图像大小为8x8，因此，预测变量个数为64个。
- 使用主成分分析模型得到的主成分因子，只需30个，使用逻辑回归模型分类，分类准确度与使用原始64个变量的准确度相差无几。
- 因此，30个主成分因子涵盖了原始64个变量的信息。

```python
from sklearn.datasets import load_digits

from sklearn.pipeline import Pipeline

from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import cross_val_score

from sklearn.linear_model import LogisticRegression
```

```python
digits = load_digits()
```

```python
digits.data.shape
```

```python
sk = StratifiedKFold(digits.target, shuffle=True)
```

```python
score = cross_val_score(LogisticRegression(), digits.data, digits.target, cv = sk)
print('mean: %.4f   std: %.4f' % (score.mean(), score.std()))
```

```python
# This dataset is way to high-dimensional. Better do PCA:
pca = PCA(n_components=30)

logis = LogisticRegression()

# Do grid search over k, n_components and C:
pipeline = Pipeline([("pca", pca), ("logis", logis)])

score2 = cross_val_score(pipeline, digits.data, digits.target, cv = sk)
print('mean: %.4f   std: %.4f' % (score2.mean(), score2.std()))
```

## 其他的降维模型

**因子分析(factor analysis)**
- 描述隐藏在一组测量到的变量中的一些更基本的，但又无法直接测量到的隐性变量 (latent variable, latent factor)。
- 比如，如果要测量学生的学习积极性(motivation)，使用课堂中的积极参与，作业完成情况，以及课外阅读时间来代表积极性。
- 将相同本质的变量归入一个因子，还可减少变量的数目。

**Truncated SVD(singular value decomposition)**
- 用于文本挖掘。
- 被应用于文档-词语矩阵，常被称为潜在语义分析（latent semantic analysis）。
- 奇异值分解模型的变种，只计算K个最大的奇异值。
- TruncatedSVD + K-Means：聚类文本文档。

**Dictionary Learning**
- 用于图像处理，如图像修补、去噪、监督学习的识别任务。

**ICA(independent component analysis)**
- 用于分离叠加的信号。

**NMF(non-negative matrix factorization)**
- 用于数据均为非负值。
- 用于图像数据和文本数据。

**LDA(latent dirichlet allocation)**
- 用于文本挖掘。
- 主题模型，从文档中发现抽象的主题。

## 数据预处理之——数据标准化

为什么数据需要标准化？

- 通过数据标准化，所有数据消除了单位的影响，可以互相进行比较。 
- 数据标准化，也消除了变量之间由于数值大小对模型的影响。
- 数据标准化是许多机器学习模型的共同要求。
- 比如核函数为RBF的支持向量机、对系数正则化的线性模型岭回归、lasso及弹性网模型，都假定数据经过标准化处理，否则模型表现很差。

数据标准化的方法：

**z-score标准化**

- 新数据=（原数据-均值）/标准差
- 新数据均值为0，标准差为1


**min-max标准化**

- 新数据=（原数据-极小值）/（极大值-极小值）
- 新数据范围在0到1之间

**max-absolute标准化**

- 新数据= 原数据/最大绝对值
- 新数据范围在-1到1之间
- 用于稀疏数据集

**稳健标准化**

- 用于数据包含异常值，这时，使用均值和标准差就不太合适。

**数据预处理的应用：以波士顿房价数据为例**

```python
from sklearn.datasets import load_boston

from sklearn.preprocessing import StandardScaler

from sklearn.cross_validation import KFold

from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LassoCV
```

```python
boston = load_boston()
```

```python
boston.data.min(axis = 0).round()
```

```python
boston.data.mean(axis = 0).round()
```

```python
boston.data.std(axis = 0).round()
```

```python
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
boston_df['Medv'] = boston.target
boston_df.plot();
```

```python
kf = KFold(n = boston.data.shape[0], n_folds = 5, shuffle = True)
```

**没有对数据标准化，直接使用KNN模型：**

```python
score = cross_val_score(KNeighborsRegressor(), boston.data, boston.target, cv = kf)
print('mean: %.4f   std: %.4f' % (score.mean(), score.std()))
```

**先对数据标准化处理，再使用KNN模型：**

```python
scaler = StandardScaler()

knn = KNeighborsRegressor()

# Do grid search over k, n_components and C:
pipeline = Pipeline([("scaler", scaler), ("knn", knn)])

score2 = cross_val_score(pipeline, boston.data, boston.target, cv = kf)
print('mean: %.4f   std: %.4f' % (score2.mean(), score2.std()))
```

从上面的模型效果对比可以看到：

- **数据是否标准化，对K最近邻算法影响较大**；
- 因为K最近邻算法要计算距离，数值大的变量对距离的影响比数值小的变量对距离的影响大；
- 因此，标准化影响KNN的效果。

**没有对数据标准化，直接使用lasso模型：**

```python
score = cross_val_score(LassoCV(), boston.data, boston.target, cv = kf)
print('mean: %.4f   std: %.4f' % (score.mean(), score.std()))
```

**先对数据标准化处理，再使用lasso模型：**

```python
scaler = StandardScaler()

lasso = LassoCV()

# Do grid search over k, n_components and C:
pipeline = Pipeline([("scaler", scaler), ("lasso", lasso)])

score2 = cross_val_score(pipeline, boston.data, boston.target, cv = kf)
print('mean: %.4f   std: %.4f' % (score2.mean(), score2.std()))
```

从上面的模型效果对比可以看到：

- **数据是否标准化，对lasso有一定影响**；
- 因为lasso要对模型的系数进行正则化，数值大的变量与数值小的变量对变量的系数有一定的影响；
- 因此，标准化同样影响lasso的效果。

**建议：拿到数据，先进行探索，发现数据取值差异较大，或单位不同，先进行标准化，再使用模型。**

## 特征选择

功能：

- 减少特征数量、降维，使模型泛化能力更强，减少过拟合
- 增强对特征和特征值之间的理解

优势：

- 提升模型的性能
- 帮助理解数据的特点、底层结构
- 这对进一步改善模型、算法都有着重要作用

方法：

**删除低方差特征**

- 假设某特征的特征值只有0和1，并且在所有输入样本中，95%的实例的该特征取值都是1（方差很小），那就可以认为这个特征作用不大。
- 如果100%都是1（方差为0），那这个特征就没意义了。

**单变量特征选择**

- 能够对每一个特征进行测试，衡量该特征和响应变量之间的关系，根据得分扔掉不好的特征。
- 对于回归和分类问题可以采用卡方检验等方式对特征进行测试。
- 这种方法比较简单，易于运行，易于理解，通常对于理解数据有较好的效果（但对特征优化、提高泛化能力来说不一定有效）；

**递归特征消除 Recursive feature elimination (RFE)**

- 反复的构建模型（如SVM或者回归模型），选出最好（最差）的特征（可以根据系数来选），把选出来的特征放到一遍，然后在剩余的特征上重复这个过程，直到所有特征都遍历了。
- 这个过程中特征被消除的次序就是特征的排序。
- 因此，这是一种寻找最优特征子集的贪心算法。
- RFE的稳定性很大程度上取决于在迭代的时候底层用哪种模型。
- 假如RFE采用普通线性回归，没有经过正则化的回归是不稳定的，那么RFE就是不稳定的；
- 假如采用的是Ridge，而用Ridge正则化的回归是稳定的，那么RFE就是稳定的。

**基于模型进行特征选择**

- 有些机器学习模型，本身就具有对特征进行打分的机制，或者很容易将其运用到特征选择任务中。
- 例如lasso(LassoCV, LassoLarsCV)模型，可用于回归模型，lasso中的alpha控制选择特征的个数，alpha越大，选择的特征越少；
- LinearSVC、逻辑回归，可用于分类模型，控制特征个数的是参数C，C越小，选择的特征越少；
- 上面基于L1的模型缺点是：当一组预测变量之间高度相关时，只会取一个变量。
- L1正则化像非正则化线性模型一样也是不稳定的，如果特征集合中具有相关联的特征，当数据发生细微变化时也有可能导致很大的模型差异。
- 可以使用对应的基于随机的模型；
- 如RandomizedLasso和RandomizedLogisticRegression模型；
- 基于决策树的特征选择，如随机森林、ExtraTreesClassifier等等。

**汇总**

- 对于理解数据、数据的结构、特点来说，单变量特征选择是个非常好的选择。尽管可以用它对特征进行排序来优化模型，但由于它不能发现冗余（例如假如一个特征子集，其中的特征之间具有很强的关联）；
- L1正则化能够生成稀疏的模型，对于选择特征子集来说非常有用；相比起L1正则化，L2正则化的表现更加稳定，由于有用的特征往往对应系数非零，因此L2正则化对于数据的理解来说很合适。由于响应变量和特征之间往往是非线性关系，可以采用basis expansion的方式将特征转换到一个更加合适的空间当中，在此基础上再考虑运用简单的线性模型。
- 随机森林是一种非常流行的特征选择方法，它易于使用，一般不需要feature engineering、调参等繁琐的步骤，并且很多工具包都提供了平均不纯度下降方法。它的两个主要问题，1是重要的特征有可能得分很低（关联特征问题），2是这种方法对特征变量类别多的特征越有利（偏向问题）。
- 特征选择在很多机器学习和数据挖掘场景中都是非常有用的。
- 在使用的时候要弄清楚自己的目标是什么，然后找到哪种方法适用于自己的任务。
- 当选择最优特征以提升模型性能的时候，可以采用交叉验证的方法来验证某种方法是否比其他方法要好。
- 当用特征选择的方法来理解数据的时候要留心，特征选择模型的稳定性非常重要，稳定性差的模型很容易就会导致错误的结论。
- 对数据进行二次采样然后在子集上运行特征选择算法能够有所帮助，如果在各个子集上的结果是一致的，那就可以说在这个数据集上得出来的结论是可信的，可以用这种特征选择模型的结果来理解数据。
