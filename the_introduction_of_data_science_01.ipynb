## 使用直方图查看连续变量的分布

当你得到一组数据，是**连续变量**，即取连续的数值，如温度、身高等，可以使用**直方图**查看数据的大致分布。

如果你得到一组数据，是**离散变量**，即仅取有限的几个数值，如性别，只取2个数值（0或者1，0代表男，1代表女），建议使用**条形图**查看。

现在，你得到一组虚构的男性身高的数据，身高属于连续变量，看下直方图吧。

```python
np.random.seed(1234)
male_height = pd.Series(np.random.normal(loc = 170, scale = 5, size = 50).round())
```

```python
sns.distplot(male_height)
plt.title('男性身高的直方图和密度曲线');
```

## 连续变量——均值

现在，根据直方图和密度曲线看到，170和175之间的身高最多，具体数值是多少呢？怎么描述呢？


用到汇总统计量了，看下平均值吧。

```python
male_height.mean()
```

```python
sns.distplot(male_height)
plt.plot(np.ones(shape = 20)*170.3, np.linspace(0, 0.11, 20), 'r--')
plt.title('男性身高的直方图、密度曲线和均值所在直线');
```

## 存在异常值的连续变量——均值

上面的均值能够很好的描述数据的集中位置，如果再加上巨人族首领的身高数据500厘米呢？

```python
male_height_new = male_height.append(pd.Series(500))
```

```python
male_height_new.mean()
```

```python
sns.distplot(male_height_new)
plt.plot(np.ones(shape = 20)*170.3, np.linspace(0, 0.11, 20), 'r--', label = '均值')
plt.plot(np.ones(shape = 20)*176.7, np.linspace(0, 0.11, 20), 'k--', label = '添加异常值后的均值')
plt.legend()
plt.title('男性身高的新旧均值对比');
```

```python
sns.distplot(male_height)
plt.plot(np.ones(shape = 20)*170.3, np.linspace(0, 0.11, 20), 'r--', label = '均值')
plt.plot(np.ones(shape = 20)*176.7, np.linspace(0, 0.11, 20), 'k--', label = '添加异常值后的均值')
plt.legend()
plt.title('男性身高的新旧均值对比');
```

## 连续变量——中位数

这就是所谓的“统计陷阱”之一，也是常说的“被平均”。

这时怎么办呢？

**中位数——稳健，不受异常值的干扰。**

中位数的计算：把数列从小到大排序，取最中间的那个数的数值。

因为涉及到排序，在以前纯手工计算的年代，计算上没有均值那么方便，求和除以总个数即可，所以应用较少。

```python
male_height_new.median()
```

```python
sns.distplot(male_height)
plt.plot(np.ones(shape = 20)*170.3, np.linspace(0, 0.11, 20), 'r--', label = '均值')
plt.plot(np.ones(shape = 20)*176.7, np.linspace(0, 0.11, 20), 'k--', label = '添加异常值后的均值')
plt.plot(np.ones(shape = 20)*171.0, np.linspace(0, 0.11, 20), 'b--', label = '添加异常值后的中位数')
plt.legend()
plt.title('男性身高的新旧均值与中位数的对比');
```

**方差为什么也被异常值影响呢？**

因为均值被异常值影响，方差是通过均值计算的。

方差的计算：
- 计算每个点到均值的距离为 Xi - mean(x)
- 因为距离有正有负，加起来正负容易抵消，因此，用距离的平方相加，(X1 - mean(x))^2 + (X2 - mean(x))^2
- 为了消除数列个数多少对方差的影响，需要除以总个数；
- 因为前面计算了平方和，导致方差的数值很大，不方便比较，因此常使用标准差，即对方差开平方。

```python
sns.distplot(male_height)
plt.plot(170.3 + np.ones(shape = 20)*4.83, np.linspace(0, 0.11, 20), 'g--')
plt.plot(170.3 - np.ones(shape = 20)*4.83, np.linspace(0, 0.11, 20), 'g--')
plt.plot(np.ones(shape = 20)*170.3, np.linspace(0, 0.11, 20), 'b--')
plt.title('男性身高的均值与一倍标准差');
```

## 连续变量——四分位距

既然中位数比均值稳健，是否可以使用中位数推出一个表示数据分散程度的稳健统计量呢？

这就是四分位距，也叫四分差。

- 假设100个数值，数据从小到大排序；
- 取排在第25位的数，即25%分位数，也称为第一四分位数；
- 再取排在第50位的数，称为第二四分位数，常称为中位数；
- 取排在第75位的数，即75%分位数，也是第三四分位数；
- 取排在第100位的数，即100%分位数，也称为第四四分位数；
- 第一四分位数与第三四分位数之间的距离，即25%分位数与75%分位数之间的距离，包含了50%的数据，称为四分位距。

观察四分位距，有个常用的图——箱形图，也称为盒型图。

## 连续变量——汇总统计量不适用的场景

前面的异常值只有一个，可以用中位数和四分位距解决。

但是，如果异常值的数据来了50个，和正常身高数据的个数一样多，这时情况如何呢？

```python
np.random.seed(1234)
giant_height = pd.Series(np.random.normal(loc = 500, scale = 5, size = 50).round())
```

```python
plt.hist(male_height, label = '正常身高', normed = True)
plt.hist(giant_height, label = '巨人身高', normed = True)
plt.legend()
plt.title('正常男性和巨人的身高直方图');
pd.DataFrame({'giant':giant_height, 'normal':male_height}).plot(kind = 'kde')
plt.title('正常男性和巨人的身高密度曲线');
```

从上面的直方图和密度曲线可以看出：

- 数据有两个集中位置，即有**两个波峰**；
- 一个在170厘米附近，一个围绕500厘米；
- 因此，汇总统计量无论是均值还是中位数都只反映一个集中位置，一个波峰；
- 即使求出均值或中位数，也不能代表真正的集中位置。

这时，就只能使用直方图或密度曲线或这两者的变形——**小提琴图**来观察数据的分布。

```python
male_big = male_height.append(giant_height)
```

```python
male_big.mean()
```

```python
male_big.median()
```

```python
sns.violinplot(male_big)
plt.title('正常男性和巨人的身高小提琴图');
```

小提琴图，实质上就是，一个被压扁的箱形图，外面加了个密度曲线，外形像小提琴而得名。

这里，为了效果明显，构建了巨人的身高。

现实中，可能是女性的身高和男性的身高，或者小朋友的身高和成年人的身高，构成两个或以上波峰的波形。

## 汇总


1. 连续变量的分布，使用直方图和密度曲线；
2. 如果连续变量只有一个波峰，且无异常值，使用均值代表集中位置，方差、标准差代表分散程度；
3. 如果连续变量只有一个波峰，有异常值，使用中位数代表集中位置，四分位距代表分散程度，箱形图可视化；
4. 如果连续变量有多个波峰，使用小提琴图可视化；
5. 离散变量的分布，使用条形图查看。
